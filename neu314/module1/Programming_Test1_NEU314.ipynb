### 1b (4 pts)
Using the text data loaded into `mystring`, please count the number of instances where 'i' preceded 'e', and 'e' preceded 'i'. Print the counts for each of these instances.

```python
## Approach 1: Vanilla Python (list comprehension)
## Iterate over pairs of characters, looking for each of the string pairs. 
## Note we set the pairs to lowercase to ensure that all instances of desired pairs
## with varying cases are counted.

i_before_e = sum( [ 'ie' == mystring[n:n+2].lower() for n in range(len(mystring)-1) ] )
e_before_i = sum( [ 'ei' == mystring[n:n+2].lower() for n in range(len(mystring)-1) ] )

print('i before e: %s' %i_before_e)
print('e before i: %s' %e_before_i)
```

```python
## Approach 2: Vanilla Python (while loop)
## Within a while loop, iterate over pairs of characters, looking for each 
## of the string pairs and count the number of iterations. Terminate loop
## once number of iterations is one less the total length of the string.
## Note we set the pairs to lowercase to ensure that all instances of 
## desired pairs with varying cases are counted.

## Define counts.
i_before_e = e_before_i = 0
n = 0

## Main loop.
while n < len(mystring)-1:
    
    ## Extract pair.
    pair = mystring[n:n+2].lower()
    
    ## Check if pair matches desired substrings.
    if 'ie' == pair: 
        i_before_e += 1
    elif 'ei' == pair:
        e_before_i += 1
    
    ## Update counter.
    n += 1
    
print('i before e: %s' %i_before_e)
print('e before i: %s' %e_before_i)
```

```python
## Approach 3: Regular expressions
## Use the regular expressions (re) package to search for instances of the desired pairs.
## Note we use the IGNORECASE flag to count all instances of the desired pairs with
## varying cases.
import re

i_before_e = len(re.findall('ie', mystring, flags=re.IGNORECASE))
e_before_i = len(re.findall('ei', mystring, flags=re.IGNORECASE))

print('i before e: %s' %i_before_e)
print('e before i: %s' %e_before_i)
```

### 1c (10 pts)
As it turns out, [i before e is a terrible rule](http://www.nathancunn.com/2017-06-26-i-before-e-except-after-w/), and this text only contains examples of violations of this rule. Let's correct them.

Using the loaded text data from **1a**:

- If an 'ei' is present **with** a 'c' preceding it, change it to **'ie'**

- If a 'ie' is present **without** a 'c' preceding it, change it to **'ei'**.

Please print the new, correct, string.

(**Hint**: use a **WHILE** loop, not a **FOR** loop)


```python
## Aproach 1: Vanilla Python (while loop)
## Within a while loop, iterate over pairs of characters and test if they match
## desired substrings. Upon a match, substitute characters as specified.

## Make a new copy of string. Convert to list for item assignment.
new_string = mystring.copy()
new_string = list(new_string)

## Main loop.
n = 0
while n < len(new_string):
    
    ## Extract pair and join as string.
    pair = ''.join(new_string[n:n+2]).lower()
    
    ## Check if pair matches desired substrings.
    ## If match, replace with desired substrings.
    if 'ei' == pair and new_string[n-1] == 'c':
        new_string[n:n+2] = 'ie'
    elif 'ie' == pair and not new_string[n-1] == 'c':
        new_string[n:n+2] = 'ei'

    ## Update counter.
    n += 1
    
## Recombine.
new_string = ''.join(new_string)
print(new_string)
```

```python
## Aproach 2: Regular Expressions
## From the regular expressions package, use the sub(stitution) function to identify
## the desired patterns and substitute any matches with the specified substrings. 
import re

## Make a new copy of string
new_string2 = mystring.copy()

## Replace all instances of 'c' preceding 'ei' with 'cie'.
new_string2 = re.sub('cei', 'cie', new_string2, flags=re.IGNORECASE)

## Replace all instances of 'ie' IF 'c' not preceding with 'ei'. This regular
## expression makes use of capture groups, which allows for the matching
## of a particular substring without its removal from the parent string. 
## For example, if we match 'fie', only the 'ie' is replaced.
new_string2 = re.sub('([^c])(ie)', r'\g<1>ei', new_string2, flags=re.IGNORECASE)
    
print(new_string2)
```

```python
## Proof of concept: Both approaches return the same string.
new_string == new_string2
```

### 3e (4 pts)
Write a function that will run the simulation $N$ times and will calculate the average interspike interval for each one of those times. Plot a single histogram of $N=100$ of these values together.

```python
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
### Define functions.
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#

def LIF_spike_isi(n_sim, V0=-0.06, dt=0.001, t_max=0.4, tau=0.02, 
                  el=-0.06, i_mean=0.025, vr=-0.07, vth=-0.05):
    '''Wrapper function to calculate the average ISI per simulation from 
    a series of spikes generated from the supplied LIF_spike function
    
    INPUTS
    - n_sim: number of simulations to run
    
    For additional arguments: see LIF_spike documentation.
    
    OUTPUTS
    - avg_isi: the mean interspike interval (ISI) for the simulation'''
    
    ## Preallocate space.
    avg_isi = np.zeros(n_sim)
    
    ## Main loop.
    for n in np.arange(n_sim):
        _, _, spike_times = LIF_spike(V0, dt, t_max, tau, el, i_mean, vr, vth)
        avg_isi[n] = np.diff(spike_times).mean()
        
    return avg_isi

#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
### Execute code.
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#

## Set RNG for replicability.
np.random.seed(47404)

## Generate distribution of mean simulated ISIs.
avg_isi = LIF_spike_isi(n_sim=100)

## Open canvas for plotting.
fig, ax = plt.subplots(1,1,figsize=(8,4))

## Plot histogram.
ax.hist(avg_isi*1e3, edgecolor='w', linewidth=0.5)

## Add details to plot.
ax.set(xlabel='Interspike Interval (ms)', ylabel='Frequency')

sns.despine()      # Remove top/right spines.
plt.tight_layout() # Tighten layout.
```

### 3g (4 pts)
Write a function `serror()` that will compute the standard error of the values in one of your histograms (i.e., the standard error of the mean ISIs, where each mean ISI is from a single simulation, and you've run $N$ simulations). As a reminder, the standard error $\sigma_e$ of a set of $N$ observations $x_i$ for $i=0...N-1$ can be found through

$$
\mu = \frac{1}{N} \sum_{i=0}^{n-1} x_i 
~ \\
~ \\
\sigma^2 = \frac{1}{N-1} \sum_{i=0}^{N-1} (x_i - \mu)^2
~ \\
~ \\
\sigma_e = \sqrt{ \frac{\sigma^2}{N} }
$$

In other words, $\mu$ is the mean, $\sigma^2$ is the variance (which is itself the square of standard deviation $\sigma$), and $\sigma_e$ is a simple function of these quantities. Feel free to use the `NumPy` functions `np.mean()` and `np.std()` in your solution.

Your function `serror()` should take in the number of simulations $N$ as a required input and the mean input current to be applied in each simulation `i_mean` as an optional input with an appropriate default value. 

Print out a call to your `serror()` function with $N=100$ and the mean input set to .03.

```python
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
### Define functions.
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#

def serror(n_sim=100, i_mean=0.025):
    '''Computes the mean and standard error across a distribution
    of mean ISIs from simulated spike data
    
    INPUTS
    - n_sim: number of simulations
    - i_mean: mean input current (default = 0.025)
    
    OUTPUTS
    - mu: average across mean ISIs
    - se: standard error across mean ISIs'''
    
    ## Simulate average ISI.
    avg_isi = LIF_spike_isi(n_sim=n_sim, i_mean=i_mean)

    ## Compute mean and standard error.
    mu = avg_isi.mean()
    sd = avg_isi.std()
    se = sd / np.sqrt(n_sim)
    return mu, se

#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
### Execute code.
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#

## Set RNG for replicability.
np.random.seed(47404)

## Compute mean and standard error if ISI for simulation.
mu, se = serror(n_sim=100, i_mean=0.03)

print('serror(N=100, i_mean=0.3): mu = %0.3f, se = %0.3f' %(mu, se))
```

### 3h (4pts)
Finally, write a function `serror_plotter()` that receives a list of values for $N$ (for example, `N = np.array([20, 40, 80, 160, 320])` ) as input and plots the standard error as a function of $N$. Again, also add the mean input current to be applied in each simulation `i_mean` as an optional input with an appropriate default value. 

What do you see? If you were doing an experiment and you wanted to reduce your error bars by a factor of 4, how many more data points $N$ would you need to collect?

```python
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
### Define functions.
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#

def serror_plotter(N, i_mean=0.025, ax=False):
    '''Convenience function: plots the standard error of mean
    ISIs, from a distribution of simulated spike datasets,
    as a function of dataset size
    
    INPUTS
    - N: list/array, number of simulations per dataset
    - i_mean: mean input current (default = 0.025)
    - ax: canvas to plot on (default = False, opens new plot)
    
    OUTPUTS
    - se: standard errors
    '''
    
    ## Iteratively estimate standard error using the
    ## serror function.
    se = [serror(n_sim=n, i_mean=i_mean)[-1] for n in N]
    
    ## Open new canvas (if applicable).
    if not ax: fig, ax = plt.subplots(1,1,figsize=(12,4))
        
    ## Plot standard errors.
    ax.plot(N, se, label=i_mean)
    ax.scatter(N, se)
    ax.set(xlabel='Sample Size', ylabel='Standard Error')
    
    return se

#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
### Execute code.
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
## NOTE: Code executes slowly. 

## Set RNG for replicability.
np.random.seed(47404)

## Define sizes of simulated datasets.
## Here we sample: 10, 100, 1000, 10000.
N = np.power(10., np.arange(1, 5)).astype(int)

## Open canvas for plotting.
fig, ax = plt.subplots(1,1,figsize=(12,4))

## Iteratively plot standard error as a function of sample size
## for several different values of input mean.
for i_mean in np.arange(0.02,0.051,0.01):

    ## Run convenience function.
    se = serror_plotter(N, i_mean=i_mean, ax=ax)
    
## Add additional information to plot.
ax.set(xscale='log', yscale='log')
ax.set(yticks=np.power(10., -np.arange(2,6)), ylabel='Standard Error', title='Log-Log Scale')

## Add legend.
legend = ax.legend(loc=7, bbox_to_anchor=(1.125,0.4), fontsize=16, borderpad=0, labelspacing=0.1)
legend.set_title('Mean Input', prop=dict(size=14))

sns.despine()      # Remove top/right spines.
plt.tight_layout() # Tighten layout.
```

As can be observed from the plot above, independent of mean input current, the standard error decreases linearly on log-log scale with sample size such that larger and larger samples are necessary to achieve similar reductions. This power law-like relationship can be observed below, where an order-of-magnitude increase in sample size achieves near 4x (closer to 3.2x) reductions in standard error. 

```python
for i in range(3):
    s = 'At i_mean = %s: from N=%s to N=%s yields SE factor increase of %0.3f.'
    print(s %(i_mean, N[i], N[i+1], se[i] / se[i+1]))
```
